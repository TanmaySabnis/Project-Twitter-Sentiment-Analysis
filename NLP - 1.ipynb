```python
# Data Cleaning for NLP
documents = []
for category in movie_reviews.categories():
    for fileID in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileID), category))
documents
```

```python
# Let's shuffle this data
import random
random.shuffle(documents)
documents
```

```python
from nltk import wordnet as wn
def get_simple_pos(tag):
    if tag.startswith("J"):
        return 'a'
    elif tag.startswith("V"):
        return 'v'
    elif tag.startswith("N"):
        return 'n'
    elif tag.startswith("R"):
        return 'r'
    else:
        return 'n'
```

```python
wn.NOUN
```

```python
# Creating stops
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
```

```python
# Function for cleaning review
lemmatizer = WordNetLemmatizer()
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w , pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
# Going through documents

documents = [(clean_review(document),category) for document,category in documents]
```

# Building a feature set

```python
training_documents = documents[0:1500]
testing_documents = documents[1500:]
```

```python
all_words = []
for document in training_documents:
    all_words += document[0]
all_words[:50]
```

```python
import nltk
freq = nltk.FreqDist(all_words)
```

```python
common = freq.most_common(3000)
features = [ele[0] for ele in common]
features
```

```python
def get_feature_dict(words):
    curr_features = {}
    words_set = set(words)
    for w in features:
        curr_features[w] = w in words_set
    return curr_features
```

```python
training_data = [(get_feature_dict(doc),category) for doc, category in training_documents]
```

```python
testing_data = [(get_feature_dict(doc),category) for doc, category in testing_documents]
```

```python
training_data[0]
```

```python
categories = [category for document, category in documents]
```

```python
text_documents = [" ".join(document) for document, category in documents]
```

```python
text_documents[0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(text_documents, categories)
```

```python
count_vec = CountVectorizer(max_features= 2000)
x_train_features = count_vec.fit_transform(X_train)
x_train_features.todense()
```

```python
x_test_features = count_vec.transform(X_test)
x_test_features.todense()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# now we have everything ready, lets try svm on it
svc = SVC()
svc.fit(x_train_features, y_train)
y_pred = svc.predict(x_test_features)
print(confusion_matrix(y_pred, y_test))
print(classification_report(y_pred, y_test))
```

#### Applying GridsearchCV for optimization

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
```

```python
svc = SVC()
grid = {'C' : [1e2, 1e3, 5e3, 1e4],
       'gamma' : [1e-3, 5e-4, 1e-4]}

grid_search = GridSearchCV( svc, grid, cv = KFold(n_splits=10, shuffle=True, random_state=1))
grid_search.fit(x_train_features, y_train)
grid_search.best_score_
```

```python
grid_search.best_estimator_
```

```python
svc = SVC(C=100.0, gamma=0.001)
svc.fit(x_train_features, y_train)
y_pred = svc.predict(x_test_features)
print(confusion_matrix(y_pred, y_test))
print(classification_report(y_pred, y_test))
```
